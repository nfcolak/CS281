```python
#import
import sqlite3
import PySimpleGUI as sg

# DB'E BAĞLANTI KURULUYOR
con = sqlite3.connect('ScooterAdminPanel.sqli')
cur = con.cursor()

# GLOBAL DEĞİŞKENLER

login_user_ssn = -1
login_user_name = -1
login_user_surname = -1
login_user_type = -1

def window_login():
    
    layout11 = [[sg.Text('Welcome to Scooter Company')],
               [sg.Text('SSN:', size=(8,1)), sg.Input(size=(8,1), key='SSN')],
               [sg.Text('Password:', size=(8,1)), sg.Input(size=(8,1), key='Password')],
               [sg.Button('Login')]]
    
    return sg.Window('Login', layout11)

def window_manager():
    
    layout1 = [[sg.Text('Welcome to Scooter Company System')],
              [sg.Button('View Repairers'), sg.Button('Add a Repairer')],
              [sg.Button('Charger Status'), sg.Button('Add a Charger')],
              [sg.Button('Add New Area'), sg.Button('View Areas'), sg.Button('Assign Charger')],
              [sg.Button('New Discount Offer'), sg.Button('View Discount Offers')],
              [sg.Button('Block Customer'), sg.Button('View Customers')],
              [sg.Button('View Warehouse Data'), sg.Button('View Parts')],
              [sg.Button('Provide Warehouse Data')],
              [sg.Button('Logout')]]
    
    return sg.Window('Manager Window', layout1)

def window_charger():
    
    layout12 = [[sg.Text('Welcome to Charger Panel')],
              [sg.Button('View Low Battery Scooters')],
              [sg.Button('Logout')]]
    
    return sg.Window('Charger Window', layout12)

def window_repairer():
    
    layout13 = [[sg.Text('Welcome to Repairer Panel')],
              [sg.Button('View Damaged Scooters')],
              [sg.Button('Logout')]]
    
    return sg.Window('Charger Window', layout13)

def window_add_repairer():
    
    layout14 = [[sg.Text('SSN: ', size=(20,1)), sg.Input(size=(20,1), key='ssn')],
               [sg.Text('Phone number: ', size=(20,1)), sg.Input(size=(20,1), key='uphoneno')],
               [sg.Text('Name: ', size=(20,1)), sg.Input(size=(20,1), key='uname')],
               [sg.Text('Surname: ', size=(20,1)), sg.Input(size=(20,1), key='usurname')],
               [sg.Text('Address: ', size=(20,1)), sg.Input(size=(20,1), key='uaddress')],
               [sg.Text('Password: ', size=(20,1)), sg.Input(size=(20,1), key='upassword')],
               [sg.Button('Add Repairer'), sg.Button('Return To Menu')]]
    
    return sg.Window('Add Repairer', layout14)

def window_add_charger():
    
    layout15 = [[sg.Text('SSN: ', size=(20,1)), sg.Input(size=(20,1), key='ssn')],
               [sg.Text('scootermo: ', size=(20,1)), sg.Input(size=(20,1), key='scootermo')],
               [sg.Text('Area Name: ', size=(20,1)), sg.Input(size=(20,1), key='area_name')],
               [sg.Text('Phone number: ', size=(20,1)), sg.Input(size=(20,1), key='uphoneno')],
               [sg.Text('Name: ', size=(20,1)), sg.Input(size=(20,1), key='uname')],
               [sg.Text('Surname: ', size=(20,1)), sg.Input(size=(20,1), key='usurname')],
               [sg.Text('Address: ', size=(20,1)), sg.Input(size=(20,1), key='uaddress')],
               [sg.Text('Password: ', size=(20,1)), sg.Input(size=(20,1), key='upassword')],
               [sg.Button('Add Charger'), sg.Button('Return To Menu')]]
    
    return sg.Window('Add Charger', layout15)

def window_view_discount_offers():
    
    discount_offers = []
    
    for row in cur.execute ('''Select d_id, CCI, vdate, dis_code
                               From discount_offer'''):
        discount_offers.append((row[0], row[1], row[2], row[3]))
    
    column_names = ['DiscountID CCI VDate Discount Code']
    
    layout16 = [[sg.Listbox(column_names + discount_offers, size=(40,10), key='discount_offers')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('View Discount Offers', layout16)

def window_view_customers():
    
    customers = []
    
    for row in cur.execute ('''Select uname, usurname
                               From User, Customer
                               Where Customer.SSN = User.SSN'''):
        customers.append((row[0], row[1]))
    
    column_names = ['Name Surname']
    
    layout17 = [[sg.Listbox(column_names + customers, size=(40,10), key='customers')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('View Customers', layout17)

def window_view_repairers():
    
    repairers = []
    
    for row in cur.execute ('''Select R_SSN
                               From Repairer'''):
        repairers.append((row[0]))
    
    repairer_infos = []
    for row in cur.execute ('''Select uname, usurname 
                               From Repairer, User
                               Where Repairer.R_SSN = User.SSN'''):
        
        repairer_infos.append((row[0], row[1]))
    print(repairer_infos)
    column_names = ['Name Surname']
    
    layout3 = [[sg.Listbox(column_names + repairer_infos, size=(40,10), key='repairer_infos')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Repairer View Window', layout3)

def window_view_parts():

    parts = []
    
    for row in cur.execute ('''Select partno, part_name, quantity_in_stock
                               From Parts'''):
        parts.append(row)
    column_names = ['Part No. Part Name Quantity']
    layout18 = [[sg.Listbox(column_names + parts, size=(40,10), key='parts')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('View Parts', layout18)

def window_view_area():

    areas = []
    
    for row in cur.execute ('''Select area_name
                               From Area'''):
        areas.append(row)
    column_names = ['Area Name']
    layout18 = [[sg.Listbox(column_names + areas, size=(40,10), key='areas')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('View Parts', layout18)
    
def window_charger_status():
    
    locations = []
    
    for row in cur.execute ('''Select area_name
                               From Charger'''):
        locations.append((row[0]))
    
    charger_status = []
    column_names = ['Name Surname Area']
    
    for row in cur.execute (''' Select uname, usurname, area_name 
                                From User, Assign
                                Where User.SSN = Assign.CHSSN'''):
        
        charger_status.append((row[0], row[1], row[2]))
    
    
    layout4 = [[sg.Listbox((column_names + charger_status), size=(40,10), key='charger_status')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Charger Status Window', layout4)


def window_new_area():
    
    layout5=[[sg.Text('Area Name: ', size=(20,1)), sg.Input( size=(20,1), key='area_name')],
             [sg.Button('Add'), sg.Button('Return To Menu')]]
                      
    return sg.Window('Add New Area', layout5)

def window_new_discount_offer():
    
    
    layout6=[[sg.Text('Discount Id: ', size=(20,1)), sg.Input( size=(20,1), key='d_id')],
             [sg.Text('CCI: ', size=(20,1)), sg.Input(key='CCI', size=(20,1))],
             [sg.Text('vdate: ', size=(20,1)), sg.Input(key='vdate', size=(20,1))],
             [sg.Text('dis_code: ', size=(20,1)), sg.Input(key='dis_code', size=(20,1))],
             [sg.Text('Discount %', size=(20,1)), sg.Input(key='discount_percentage', size=(20,1))],
             [sg.Button('Add Discount Offer'), sg.Button('Return To Menu')]]
                      
    return sg.Window('Add New Discount Offer', layout6)

def window_block_customer():
    
    customers = []
    
    for row in cur.execute ('''Select uname, usurname
                               From User, Customer
                               Where Customer.SSN = User.SSN'''):
        customers.append((row[0], row[1]))
        
    layout7 = [[sg.Text('Customer: '), sg.Combo(customers, size=(40,20), key='customer')],
              [sg.Button('Block')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Block', layout7)

def window_view_warehouse_data():
    warehouse_data = []
    
    for row in cur.execute ('''Select address, capacity
                               From Warehouse'''):
        warehouse_data.append(row)
    column_names = ['Address  Capacity']
    layout8 = [[sg.Listbox(column_names + warehouse_data, size=(40,10), key='warehouse_data')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('View Warehouse Data', layout8)

def window_provide_warehouse_data():
    
    parts = []
    column_names = ['Part No. Part Name Quantity']
    
    for row in cur.execute ('''Select partno, part_name, quantity_in_stock
                               From Parts'''):
        parts.append((row[0], row[1], row[2]))
        
    layout = [[sg.Listbox(column_names + parts, size=(60,10), key='parts')],
             [sg.Text('Part Name: ', size=(20,1)), sg.Input(size=(10,1), key='part_name'), sg.Button('Update Part Name')],
             [sg.Text('Quantity: ', size=(20,1)), sg.Input(size=(10,1), key='quantity'), sg.Button('Update Quantity')],
             [sg.Button('Return To Menu')]]
    
    return sg.Window('Provide Data', layout)

def button_update_part_name(values):
    parts = values['parts']
    new_part_name = values['part_name']
    
    if len(parts) == 0:
        sg.popup('Choose a part')
    elif new_part_name == '':
        sg.popup('You cannot leave new part name empty')
    else:
        cur.execute('''Update Parts
                       set part_name = ? Where partno = ?''', (new_part_name, parts[0][0]))
        sg.popup('Done')
        window.Element('part_name').Update(value = '')
        
def button_update_quantity(values):
    
    parts = values['parts']
    new_quantity = values['quantity']
    
    if len(parts) == 0:
        sg.popup('Choose a part')
    elif new_quantity == '':
        sg.popup('You cannot leave new quantity empty')
    elif not new_quantity.isnumeric():
        sg.popup('Quantity cannot be non-numeric')
    else:
        cur.execute('''Update Parts
                       set quantity_in_stock = ? Where partno = ?''', (new_quantity, parts[0][0]))
        sg.popup('Done')
        window.Element('quantity').Update(value = '')
        
def window_assign_charger():
    
    chargers = []
    areas = []
    
    for row in cur.execute ('''Select area_name
                               From Area'''):
        areas.append((row[0]))
     
    for row in cur.execute (''' Select uname, usurname, User.SSN
                                From Charge, User
                                Where User.SSN = Charge.CHSSN'''):
        
        chargers.append((row[0], row[1], row[2]))
        
    layout = [[sg.Listbox(areas, size=(10,10), key='areas'), sg.Listbox(chargers, size=(30,10), key='chargers')],
              [sg.Button('Assign the Charger'), sg.Button('Return To Menu')]]
    
    return sg.Window('Assign Charger', layout)

def button_assign_charger(values):
    
    areas = values['areas']
    chargers = values['chargers']
    m_id = []
    print(areas)
    print(chargers)
    
    if len(areas) == 0:
        sg.popup('Please select an area')
    elif len(chargers) == 0:
        sg.popup('Please select a charger')
    else:
        for row in cur.execute('''Select m_id
                                  From Assign Order By m_id DESC LIMIT 1'''):
            m_id.append((row[0]))
            
        cur.execute('''Select CHSSN
                       From Assign Where CHSSN = ?''', (chargers[0][2],))
        row = cur.fetchone()
        
        if row is None:
            cur.execute('''Insert Into Assign VALUES(?,?,?)''', (m_id[0] + 1, chargers[0][2], areas[0][0]))
            sg.popup('Charger assigned')
        else:
            sg.popup('Charger is occupied')
            
def window_view_low_battery_scooters():

    locations = []
    scooters = []
    
    for row in cur.execute(''' Select Location
                               From Scooter'''):
        locations.append((row[0]))
        
    scooter_infos = []
    
    for row in cur.execute(''' Select SID, CCI, cost, hours, battlevel, Location
                               From Scooter'''):
        scooter_infos.append((row[0], row[1], row[2], row[3], row[4], row[5]))
    
    layout14 = [[sg.Text('Locations: '), sg.Combo(locations, size=(40,20), key='locations'), sg.Button('List Low Battery Scooters')],
              [sg.Listbox(scooters, size=(40,10), key='scooters')],
              [sg.Button('Charge Scooter')],
              [sg.Button('Request Charged Battery')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Available Scooters', layout14)

def window_view_damaged_scooters():
    
    locations = []
    scooters = []
    parts = []
    scooter_infos = []
    
    for row in cur.execute('''Select part_name
                              From Parts'''):
        parts.append((row[0]))
    
    for row in cur.execute('''Select Location
                              From Scooter'''):
        locations.append((row[0]))
    
    for row in cur.execute('''Select SID, Location
                              From Scooter'''):
        scooter_infos.append((row[0], row[1]))
    
    layout15 = [[sg.Text('Locations: '), sg.Combo(locations, size=(40,20), key='locations'), sg.Button('List Damaged Scooters')],
              [sg.Listbox(scooters, size=(40,10), key='scooters'), sg.Listbox(parts, size=(40,10), key='parts')],
              [sg.Button('Fix Scooter')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Damaged Scooters', layout15)
 

def button_add_new_area(values):
    area_name = values['area_name']
    
    if area_name == '':
        sg.popup('Area name cannot be empty')
        
    else:
        cur.execute ('''Select area_name
                        From Area Where area_name=?''', (area_name,))
        row = cur.fetchone()
        
        if row is None:
            cur.execute('Insert Into Area VALUES(?)', (area_name,))
            
            sg.popup('Successfully added ' + area_name)
            
            window.Element('area_name').Update(value='')
        else:
            sg.popup('This area already exists in the system')
            window.Element('area_name').Update(value='')
            

def button_add_new_discount_offer(values):
    discountId = values['d_id']
    cci = values['CCI']
    valDate = values['vdate']
    discountCode = values['dis_code']
    discount_percentage = values['discount_percentage']
    
    if discountId == '':
        sg.popup('Discount ID cannot be empty')
        
    elif cci == '':
        sg.popup('CCI cannot be empty')
        
    elif valDate == '':
        sg.popup('Validation date cannot be empty')
        
    elif discountCode == '':
        sg.popup('Discount code cannot be empty')
        
    elif discount_percentage == '':
        sg.popup('Discount percentage code cannot be empty')
          
    else:
        cur.execute ('''Select dis_code
                        From discount_offer Where dis_code=?''', (discountCode,))
        row = cur.fetchone()
        
        if row is None:
            cur.execute('Insert Into discount_offer VALUES(?,?,?,?,?)', (discountId, cci, valDate, discountCode, discount_percentage))
            
            sg.popup('Successfully added!')
            
            window.Element('d_id').Update(value='')
            window.Element('CCI').Update(value='')
            window.Element('vdate').Update(value='')
            window.Element('dis_code').Update(value='')
            window.Element('discount_percentage').Update(value='')

def button_block_customer(values):
    customer = values['customer']

    if customer == '':
        sg.popup('Choose a customer to block')
    else:
        cur.execute('''DELETE From User Where User.uname=? and User.usurname = ?''', (customer[0], customer[1]))
        sg.popup('Customer was successfully blocked')

def window_customer_info():
    
    layout = [[sg.Text('Welcome ' + login_user_name + ' ' + login_user_surname)],
              [sg.Text('Address: ', size=(20,1)), sg.Input( size=(20,1), key='address')],
              [sg.Text('Phone number: ', size=(20,1)), sg.Input(key='phone_number', size=(20,1))],
              [sg.Button('Continue'), sg.Button('Logout')]]
                      
    return sg.Window('Customer Info', layout)

def button_continue_to_window_customer(values):
    
    global window
    address = values['address']
    phone_number = values['phone_number']
    
    if address == '':
        sg.popup('Address cannot be empty')
        
    elif phone_number == '':
        sg.popup('Phone number cannot be empty')
        
    elif not phone_number.isnumeric():
        sg.popup('Phone number should be a number')
        
    else:
        sg.popup('Thank you for using Scooter Company')
        window.close()
        window = window_customer()

def window_customer():
    
    layout9 = [[sg.Text('Welcome ' + login_user_name + ' ' +login_user_surname)],
             [sg.Button('View Scooters Available In My Area')], #Also rent a scooter (maybe use a discount offer)
             [sg.Button('Logout')]]
    
    return sg.Window('Customer Window', layout9)

def window_view_available_scooters():
    
    locations = []
    scooters = []
    
    for row in cur.execute(''' Select Location
                               From Scooter'''):
        locations.append((row[0]))
        
    scooter_infos = []
    
    for row in cur.execute(''' Select SID, CCI, cost, hours, battlevel, Location
                               From Scooter'''):
        scooter_infos.append((row[0], row[1], row[2], row[3], row[4], row[5]))
        
    discount_offers = []
    
    for row in cur.execute(''' Select discount_percentage
                               From discount_offer'''):
        discount_offers.append((row[0]))
        
    layout10 = [[sg.Text('Locations: '), sg.Combo(locations, size=(40,20), key='locations'), sg.Button('List Available Scooters')],
              [sg.Listbox(scooters, size=(40,10), key='scooters')],
              [sg.Text('Discount Offers: '), sg.Combo(discount_offers, size=(40,20), key='discount_offers')],
              [sg.Button('Rent')],
              [sg.Button('Return To Menu')]]
    
    return sg.Window('Available Scooters', layout10)

def button_list_available_scooters(values):
    locations = values['locations']
    scooters = []
    
    for row in cur.execute('''Select SID, CCI, cost, hours, battlevel, Location, Availability
                              From Scooter
                              Where Availability = ? and Location = ? and Condition = ?''', ('Available',locations,'Not Damaged')):
    
        scooters.append((row[0], row[1], row[2], row[3], row[4], row[5], row[6]))
        
    window.Element('scooters').Update(values = scooters)

def button_rent(values):
    
    locations = values['locations']
    scooters = values['scooters']
    discount_offers = values['discount_offers']
    
    if len(locations) == 0:
        sg.popup('Please choose a location')
    elif len(scooters) == 0:
        sg.popup('Choose a scooter')        
    else:
        cur.execute('''UPDATE Scooter set Availability = ? Where Scooter.SID = ?''', ('Not Available', scooters[0][0]))
        row = cur.fetchone()
        
        sg.popup('Scooter was rented')
        
        window.Element('scooters').Update(values = '')

def button_login(values):
    
    global login_user_ssn
    global login_user_name
    global login_user_surname
    global login_user_type
    global window
    
    ussn = values['SSN']
    upass = values['Password']
    
    if ussn == '' and upass == '':
        sg.popup('SSN and Password cannot be empty!')
    elif ussn =='':
        sg.popup('SSN cannot be empty!')
    elif upass=='':
        sg.popup('Password cannot be empty!')
    else:
        cur.execute('Select SSN, uname, usurname From User Where SSN= ? AND upassword = ?', (ussn, upass))
        row = cur.fetchone()
        
        if row is None:
            sg.popup('SSN or password is wrong!')
        else:
            login_user_ssn = row[0]
            
            login_user_name = row[1]
            
            login_user_surname = row[2]

            cur.execute('Select SSN FROM Customer Where SSN = ?', (ussn,))
            row_customer = cur.fetchone()
            
            if row_customer is None:
                
                cur.execute('Select SSN FROM Manager Where SSN = ?', (ussn,))
                row_manager = cur.fetchone()
                
                if row_manager is None:
                    
                    cur.execute('Select SSN FROM Charger Where SSN = ?', (ussn,))
                    row_charger = cur.fetchone()
                    
                    if row_charger is None:
                        
                        cur.execute('Select R_SSN FROM Repairer Where R_SSN = ?', (ussn,))
                        row_repairer = cur.fetchone()
                    
                        if row_repairer is None:
                            sg.popup('User type error! Please contact the admin.')
                    
                        else:
                            login_user_type = 'Repairer'
                            sg.popup ('Welcome, ' + login_user_name + ' ' +login_user_surname + ' (Repairer)')
                            window.close()
                            window = window_repairer()
                    else: 
                        login_user_type = 'Charger'
                        sg.popup ('Welcome, ' + login_user_name + ' ' + login_user_surname + ' (Charger)')
                        window.close()
                        window = window_charger()
                else:
                    login_user_type = 'Manager'
                    sg.popup ('Welcome, ' + login_user_name + ' ' + login_user_surname + ' (Manager)')
                    window.close()
                    window = window_manager()
            else:
                login_user_type = 'Customer'
                sg.popup ('Welcome, ' + login_user_name + ' ' + login_user_surname + ' (Customer)')
                window.close()
                window = window_customer_info()  

def button_list_low_battery_scooters(values):
    
    locations = values['locations']
    scooters = []
    column_names = ['SID  Bat. Level  Location']
    for row in cur.execute ('''Select SID, battlevel, Location
                               From Scooter
                               Where Location = ? and battlevel <= 30''', (locations,)):
        
        scooters.append((row[0], row[1], row[2]))
    if not scooters:
        sg.popup('There is no scooter with low battery at this address')
    else:
        window.Element('scooters').Update(values = column_names + scooters)
    
def button_charge_scooter(values):
    
    locations = values['locations']
    scooters = values['scooters']

    if len(scooters) == 0:
        sg.popup('Choose a scooter to charge')
    else:
        cur.execute('''UPDATE Scooter set battlevel = 100 Where Scooter.SID = ?''', (scooters[0][0],))
        row = cur.fetchone()
        
        sg.popup('Battery was charged')
        
        window.Element('scooters').Update(values = '')
    
def button_request_charged_battery(values):
    
    scooters = values['scooters']
    battery_stocks = []
    battery_quantity = []

    if len(scooters) == 0:
        sg.popup('Choose a scooter to change its battery')
    else:
        for row in cur.execute('''SELECT partno
                              FROM Battery Order By partno DESC LIMIT 1'''):
            battery_stocks.append((row[0]))  
            
        for row in cur.execute('''SELECT quantity_in_stock
                                  FROM Parts WHERE partno = ?''', (battery_stocks[0],)):
            battery_quantity.append((row[0]))
 
        cur.execute('''UPDATE Parts set quantity_in_stock = ? Where partno = ?''', (battery_quantity[0] - 1, battery_stocks[0]))

        cur.execute('''DELETE FROM Battery WHERE b_id = (SELECT min(b_id) FROM Battery)''')
        
        cur.execute('''UPDATE Scooter set battlevel = 100 Where Scooter.SID = ?''', (scooters[0][0],))

        sg.popup('Battery requested')

def button_list_damaged_scooters(values):
 
    locations = values['locations']
    scooters = []
    column_names = ['SID  Location']
    
    for row in cur.execute ('''Select SID, Location
                               From Scooter
                               Where Location = ? and Condition = ?''', (locations,'Damaged')):
        
        scooters.append((row[0], row[1]))
    if not scooters:
        sg.popup('There is no damaged scooter at this address')
    else:
        window.Element('scooters').Update(values = column_names + scooters)
window = window_login()

def button_fix_scooter(values):

    scooters = values['scooters']
    parts = values['parts']
    stocks = []
    quantity = []
    
    if len(scooters) == 0:
        sg.popup('Choose a scooter to fix')
    elif len(parts) == 0:
        sg.popup('Choose a part to request')
    else:
        for row in cur.execute('''SELECT partno
                                  FROM Parts 
                                  Where part_name = ? 
                                  Order By partno DESC LIMIT 1''', (parts[0],)):
            stocks.append((row[0]))  
            
        print(parts)    
        
        for row in cur.execute('''SELECT quantity_in_stock
                                  FROM Parts WHERE partno = ?''', (stocks[0],)):
            quantity.append((row[0]))
        
        cur.execute('''UPDATE Scooter set Condition = ? Where Scooter.SID = ?''', ('Not Damaged', scooters[0][0]))
        row = cur.fetchone()
        
        cur.execute('''UPDATE Parts set quantity_in_stock = ? Where partno = ?''', (quantity[0] - 1, stocks[0]))
        
        sg.popup('Scooter was fixed')
        
        window.Element('scooters').Update(values = '')
        
def button_add_repairer(values):
    
    ssn = values['ssn']
    uphoneno = values['uphoneno']
    uname = values['uname']
    usurname = values['usurname']
    uaddress = values['uaddress']
    upassword = values['upassword']

    if ssn == '':
        sg.popup('Fill in SSN')
    else:
        cur.execute('''INSERT INTO Repairer VALUES(?)''', (ssn,))
        cur.execute('''INSERT INTO User VALUES(?,?,?,?,?,?)''', (ssn, uphoneno, uname, usurname, uaddress, upassword))

        sg.popup('Successful')
        
def button_add_charger(values):
    
    ssn = values['ssn']
    scootermo = values['scootermo']
    area_name = values['area_name']
    uphoneno = values['uphoneno']
    uname = values['uname']
    usurname = values['usurname']
    uaddress = values['uaddress']
    upassword = values['upassword']

    if ssn == '':
        sg.popup('Fill in SSN')
    else:
        cur.execute('''INSERT INTO Charger VALUES(?,?,?)''', (ssn, scootermo, area_name))
        cur.execute('''INSERT INTO User VALUES(?,?,?,?,?,?)''', (ssn, uphoneno, uname, usurname, uaddress, upassword))

        sg.popup('Successful')
        
while True:
    event, values = window.read()
    
    if event == 'Login':
        button_login(values)
    elif event == 'Continue':
        button_continue_to_window_customer(values)
    elif event == 'View Repairers':
        window.close()
        window = window_view_repairers()
    elif event == 'Add a Repairer':
        window.close()
        window = window_add_repairer()
    elif event == 'Add Repairer':
        button_add_repairer(values)
    elif event == 'Add a Charger':
        window.close()
        window = window_add_charger()
    elif event == 'Add Charger':
        button_add_charger(values)
    elif event == 'Charger Status':
        window.close()
        window = window_charger_status()
    elif event == 'Assign Charger':
        window.close()
        window = window_assign_charger()
    elif event == 'Assign the Charger':
        button_assign_charger(values)
    elif event == 'View Discount Offers':
        window.close()
        window = window_view_discount_offers()
    elif event == 'View Customers':
        window.close()
        window = window_view_customers()
    elif event == 'View Parts':
        window.close()
        window = window_view_parts()
    elif event == 'Add New Area':
        window.close()
        window = window_new_area()
    elif event == 'View Areas':
        window.close()
        window = window_view_area()
    elif event == 'Add':
        button_add_new_area(values)
    elif event == 'New Discount Offer':
        window.close()
        window = window_new_discount_offer()
    elif event == 'Add Discount Offer':
        button_add_new_discount_offer(values)
    elif event == 'Block Customer':
        window.close()
        window = window_block_customer()
    elif event == 'Block':
        button_block_customer(values)
    elif event == 'View Warehouse Data':
        window.close()
        window = window_view_warehouse_data()
    elif event == 'Provide Warehouse Data':
        window.close()
        window = window_provide_warehouse_data()
    elif event == 'Update Part Name':
        button_update_part_name(values)
    elif event == 'Update Quantity':
        button_update_quantity(values)
    elif event == 'View Scooters Available In My Area':
        window.close()
        window = window_view_available_scooters()
    elif event == 'List Available Scooters':
        button_list_available_scooters(values)
    elif event == 'Rent':
        button_rent(values)
    elif event == 'View Low Battery Scooters':
        window.close()
        window = window_view_low_battery_scooters()
    elif event == 'List Low Battery Scooters':
        button_list_low_battery_scooters(values)
    elif event == 'Charge Scooter':
        button_charge_scooter(values)
    elif event == 'Request Charged Battery':
        button_request_charged_battery(values)
    elif event == 'View Damaged Scooters':
        window.close()
        window = window_view_damaged_scooters()
    elif event == 'List Damaged Scooters':
        button_list_damaged_scooters(values)
    elif event == 'Fix Scooter':
        button_fix_scooter(values)
    elif event == 'Logout':
        login_user_ssn = -1
        login_user_name = -1
        login_user_surname = -1
        login_user_type = -1
        window.close()
        window = window_login()
    elif event == 'Return To Menu':
        if login_user_type == 'Customer':
            window.close()
            window = window_customer_info()
        elif login_user_type == 'Manager':
            window.close()
            window = window_manager()
        elif login_user_type == 'Charger':
            window.close()
            window = window_charger()
        elif login_user_type == 'Repairer':
            window.close()
            window = window_repairer()
    elif event == sg.WIN_CLOSED:
        break
        
window.close()
con.close()    
```
